## Making Popularity System

Will show the 50 most highly rated books having more than 250 reviews.

```python
books_ratings = pd.merge(books,ratings,on = 'ISBN')
```

```python
num_ratings = books_ratings.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_ratings.rename(columns = {'Book-Rating' : 'n_ratings'}, inplace = True)
```

```python
num_ratings
```

```python
avg_ratings = books_ratings.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_ratings.rename(columns = {'Book-Rating' : 'Avg_Rating'}, inplace = True)
```

```python
popularity = pd.merge(num_ratings,avg_ratings, on = 'Book-Title', how = 'inner')
```

```python
popular_df = popularity[popularity['n_ratings']>=250].sort_values(by = 'Avg_Rating', ascending= False).head(50)
```

```python
popular_df = popular_df.merge(books,on = 'Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','n_ratings','Avg_Rating']]
```

## Collaborative Filtering Based Recommendation System

Will consider users which have reviewed more than certain number of reviews and books which have received more than certain number of ratings for recommendation system.

```python
user_ratings_thr = 200
books_rating_thr = 50
```

```python
x = books_ratings.groupby('User-ID')['Book-Rating'].count() > user_ratings_thr
active_users = x[x].index
```

```python
filtered_rating = books_ratings[books_ratings['User-ID'].isin(active_users)]
```

```python
y = filtered_rating.groupby('Book-Title')['Book-Rating'].count() > books_rating_thr
famous_books = y[y].index
```

```python
filtered_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
filtered_rating = filtered_rating[['User-ID','ISBN', 'Book-Title', 'Book-Author', 'Book-Rating','Year-Of-Publication', 'Publisher',
       'Image-URL-S', 'Image-URL-M', 'Image-URL-L' ]]
```

```python
filtered_rating.duplicated().sum()
```

```python
pt = filtered_rating.pivot_table(index = 'Book-Title', columns = 'User-ID',values= 'Book-Rating')
```

```python
pt.shape
```

```python
pt
```

```python
pt = pt.fillna(0)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
def recommend(book_name):
    book_index = np.where(pt.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[book_index])),key = lambda x : x[1], reverse=True)[1:6]
    for i in similar_items:
        print(pt.index[i[0]])
```
